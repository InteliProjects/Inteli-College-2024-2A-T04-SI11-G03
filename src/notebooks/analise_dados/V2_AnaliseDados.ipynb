## Análise de correlação das colunas

A matriz de correlação é uma ferramenta estatística que mostra a relação entre diferentes variáveis. Cada célula da matriz contém o coeficiente de correlação entre um par de variáveis, que pode variar de -1 a +1. Esse coeficiente indica a força e a direção da relação linear entre as variáveis:

* +1: Correlação positiva perfeita (quando uma variável aumenta, a outra também
aumenta).
* -1: Correlação negativa perfeita (quando uma variável aumenta, a outra diminui).
* 0: Nenhuma correlação linear.

Conclusões
Para o modelo preditivo que será desenvolvido, a matriz de correlação pode ajudar a selecionar as variáveis que entram ou não no modelo. Nessa plotagem, especificamente, as correlações positivamente positivas, em maioria, são entre variáveis de tempo diretamente relacionadas, como por exemplo ‘TEMPOATENDIMENTO’ e ‘DATAPROGRAMACAO’. Além disso, para facilitar o processamento dos dados nesse notebook, a primeira versão da matriz auxiliou na escolha das variáveis que seriam retiradas, sendo elas:
* NMCOMUNIDADE
* AREAATUACAO
* TEMPOPADRAO
* TEMPOATENDIMENTO
* DT_LIMITE_EXECUCAO
* FL_PROGRAMACAO_AUTOMATICA
* EQP
* CODEXECOCOR
* SF
* NR_ROTA
* NM_TIPO_EXECUCAO
* DATAEMISSAO
* DATAPROGRAMACAO
* DATAINICIO




```python
def analise_correlacao(df):
    df_numeric = df.select_dtypes(include=['float64', 'int64', 'datetime64'])

    df_numeric = df_numeric.apply(lambda col: col.astype('int64') if col.dtype == 'datetime64[ns]' else col)

    matriz_correlacao = df_numeric.corr().round(2)

    fig = px.imshow(matriz_correlacao,
                    text_auto=True,
                    aspect="auto",
                    color_continuous_scale='RdBu_r',
                    title='Matriz de Correlação entre Variáveis')

    fig.show()

    return matriz_correlacao

correlacao = analise_correlacao(df_fraude_histo)
```

## Levantamento de hipótese
O levantamento de hipóteses é uma etapa fundamental na análise de dados, pois orienta todo o processo de investigação e tomada de decisões.

1. Há uma variação significativa nas ocorrências  ao longo dos meses em um ano.
Especula-se que pode haver picos de ocorrência em meses específicos, devido a fatores sazonais ou operacionais.

  * Conclusão: Em concordância com o que foi dito pelo parceito durante a review, nota-se que durante os anos, existem momentos específicos que o consumo de água é reduzido. Mesmo sendo o dataset de fraude, é importante considerar aspectos como férias escolares, intervalo comercial, desemprego, pandemia de Covid-19.



```python
def analise_descricao_por_ano(df):

    if 'DESCRICAO' not in df.columns or 'ANOMES' not in df.columns:
        raise ValueError("As colunas 'DESCRICAO' ou 'ANOMES' não foram encontradas no DataFrame.")

    if not pd.api.types.is_datetime64_any_dtype(df['ANOMES']):
        df['ANOMES'] = pd.to_datetime(df['ANOMES'] + '01', format='%m/%y%d', errors='coerce')

    df['Ano'] = df['ANOMES'].dt.year
    df['Mes'] = df['ANOMES'].dt.month

    df_contagem = df.groupby(['Ano', 'Mes', 'DESCRICAO']).size().reset_index(name='count')


    anos = df['Ano'].unique()

    for ano in sorted(anos):
        df_ano = df_contagem[df_contagem['Ano'] == ano]

        fig = px.line(df_ano,
                      x='Mes',
                      y='count',
                      color='DESCRICAO',
                      title=f'Quantidade de Ocorrências em {ano}',
                      labels={'count': 'Quantidade', 'Mes': 'Mês', 'DESCRICAO': 'Descrição'},
                      markers=True)

        fig.update_layout(xaxis_title='Mês',
                          yaxis_title='Quantidade',
                          xaxis={'tickvals': list(range(1, 13)), 'ticktext': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']},
                          xaxis_tickangle=-45)

        fig.show()

analise_descricao_por_ano(df_fraude_histo)

```

```python
def combine_consumo_dfs(dfs_consumo, chunksize=50000):
    """Combine os DataFrames de consumo em blocos menores para evitar problemas de memória."""
    combined_df = pd.DataFrame()

    for i, df in enumerate(dfs_consumo):
        print(f"Processando DataFrame {i + 1}/{len(dfs_consumo)} com {df.shape[0]} linhas...")

        num_chunks = (df.shape[0] // chunksize) + 1
        for start_row in range(0, df.shape[0], chunksize):
            end_row = min(start_row + chunksize, df.shape[0])
            chunk = df.iloc[start_row:end_row]

            combined_df = pd.concat([combined_df, chunk], axis=0, ignore_index=True)

    return combined_df

dfs_consumo = [df_consumo_2019, df_consumo_2020, df_consumo_2021, df_consumo_2022, df_consumo_2023, df_consumo_2024]

df_consumo_combined = combine_consumo_dfs(dfs_consumo)

print(df_consumo_combined.head())
print(f"DataFrame combinado tem {df_consumo_combined.shape[0]} linhas e {df_consumo_combined.shape[1]} colunas.")


```

## Levantamento de hipóteses


  1. O padrão de consumo de água varia ao longo do tempo e pode indicar comportamento fraudulento.

        *  Conclusão: A partir do gráfico que mostra o volume de água gasto por fraudadores ao longo dos anos, podemos observar que houve flutuações significativas nesse consumo. O pico de consumo ocorre em 2022, seguido por uma queda em 2023 e 2024.
        
        A variação nos volumes pode indicar que os fraudadores estão ajustando seu comportamento em resposta a fatores externos, como maior vigilância, penalidades mais severas ou mudanças na economia que influenciam o uso de recursos.






```python
def visualizar_volume_fraude_anos(df_consumo_combined, df_fraude_histo):

    df_fraude_matriculas = df_fraude_histo['MATRICULA']
    df_consumo_fraude = df_consumo_combined[df_consumo_combined['MATRICULA'].isin(df_fraude_matriculas)]

    df_consumo_fraude = df_consumo_fraude.merge(df_fraude_histo[['MATRICULA', 'ANOOS']], on='MATRICULA', how='left')

    df_volume_por_ano = df_consumo_fraude.groupby('ANOOS')['CONS_MEDIDO'].sum().reset_index()

    fig = px.line(df_volume_por_ano, x='ANOOS', y='CONS_MEDIDO',
                  title='Volume de Água Gasto pelos Fraudadores ao Longo dos Anos',
                  labels={'ANOOS': 'Ano', 'CONS_MEDIDO': 'Volume de Água (m3)'},
                  markers=True)
    fig.show()

visualizar_volume_fraude_anos(df_consumo_combined, df_fraude_histo)
```

  2.  Certas categorias estão mais associadas a fraudes, com maior uso de volume, em comparação com outras.

    * Conclusões: A mediana de consumo mais alta observada na categoria pública, com 25 m³, é significativa. Este resultado pode indicar que fraudes em estabelecimentos públicos tendem a envolver volumes maiores de água, possivelmente devido ao tamanho das instalações ou maior número de usuários. Essa alta mediana sugere um potencial ponto crítico, onde as fraudes podem estar passando despercebidas por longos períodos ou serem subestimadas em termos de impacto. A segunda maior mediana, na categoria residencial (8 m³), destaca a vulnerabilidade dessa categoria à fraude. Mesmo que o volume seja menor comparado à categoria pública, o alto número de unidades residenciais em uma cidade pode significar um impacto significativo no agregado. As categorias industrial e comercial apresentam medianas de consumo mais baixas, 4 m³ e 3 m³, respectivamente. Isso pode indicar que, nessas categorias, a fraude é menos comum ou ocorre em menor escala. No entanto, pode também refletir um controle mais rigoroso e mecanismos de detecção mais eficientes nessas áreas, possivelmente devido ao maior controle corporativo ou auditorias mais frequentes.



```python
def visualizar_mediana_por_categoria(df_consumo_combined, df_fraude_histo):
    df_fraude_matriculas = df_fraude_histo['MATRICULA']
    df_consumo_fraude = df_consumo_combined[df_consumo_combined['MATRICULA'].isin(df_fraude_matriculas)]

    df_mediana_por_categoria = df_consumo_fraude.groupby('CATEGORIA')['CONS_MEDIDO'].median().reset_index()

    fig = px.bar(df_mediana_por_categoria, x='CATEGORIA', y='CONS_MEDIDO',
                 title='Mediana do Volume de Água por Categoria (Fraude)',
                 labels={'CATEGORIA': 'Categoria', 'CONS_MEDIDO': 'Mediana do Consumo (m³)'},
                 text='CONS_MEDIDO')

    fig.show()

visualizar_mediana_por_categoria(df_consumo_combined, df_fraude_histo)
```

  3. O consumo de água varia de acordo com a localização geográfica, o que pode ajudar a identificar padrões setoriais de fraude.

      * Conclusões: Essa hipótese se alinha diretamente com a preocupação do cliente expressada no primeiro encontro, onde foi mencionada a necessidade de uma visibilidade mais granular e específica sobre os indivíduos ou residências. A ideia é utilizar a variação no consumo de água conforme a localização geográfica para identificar padrões setoriais de fraude. Isso permitiria uma análise mais precisa e direcionada, facilitando a classificação de um consumo como fraudulento ou não, com base em características específicas de cada área ou setor. Esse enfoque local pode revelar tendências ou anomalias que, de outra forma, poderiam passar despercebidas em uma análise mais geral.

      Uma análise interessante é a  proximidade de fraudes com grandes volumes em determinadas infraestruturas, como a Base Aérea de Campo Grande e grandes vias, sugerindo que fatores geográficos e de infraestrutura podem influenciar ou facilitar práticas fraudulentas.

```python
def visualizar_fraudes_geograficas_com_volume(df_consumo_combined, df_fraude_histo, city_coords, sample_size=10000):

    df_fraude_matriculas = df_fraude_histo['MATRICULA']
    df_consumo_fraude = df_consumo_combined[df_consumo_combined['MATRICULA'].isin(df_fraude_matriculas)]

    df_consumo_fraude_sample = df_consumo_fraude.sample(n=min(sample_size, len(df_consumo_fraude)), random_state=1)

    df_cidade_fraude = df_consumo_fraude_sample[
        (df_consumo_fraude_sample['COD_LATITUDE'] >= city_coords['lat_min']) &
        (df_consumo_fraude_sample['COD_LATITUDE'] <= city_coords['lat_max']) &
        (df_consumo_fraude_sample['COD_LONGITUDE'] >= city_coords['lon_min']) &
        (df_consumo_fraude_sample['COD_LONGITUDE'] <= city_coords['lon_max'])
    ]

    if df_cidade_fraude.empty:
        print("Nenhum dado de fraude encontrado na área filtrada. Ajuste as coordenadas ou verifique os dados.")
        return

    fig = px.scatter_mapbox(df_cidade_fraude,
                            lat='COD_LATITUDE',
                            lon='COD_LONGITUDE',
                            size='CONS_MEDIDO',
                            hover_name='MATRICULA',
                            title='Localização das Fraudes por Latitude e Longitude - Campo Grande',
                            mapbox_style="open-street-map",
                            color_discrete_sequence=['purple'])

    fig.update_layout(mapbox_center={"lat": df_cidade_fraude['COD_LATITUDE'].mean(),
                                     "lon": df_cidade_fraude['COD_LONGITUDE'].mean()},
                      mapbox_zoom=12)

    fig.show()

city_coords = {
    'lat_min': -20.5,
    'lat_max': -20.3,
    'lon_min': -54.8,
    'lon_max': -54.5
}

visualizar_fraudes_geograficas_com_volume(df_consumo_combined, df_fraude_histo, city_coords)
```
